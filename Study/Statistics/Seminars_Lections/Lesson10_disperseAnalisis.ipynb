Даны веса пациентов до и после диеты. Веса распределены нормально
До 92.8 , 95.6, 92.1, 100.6, 96.2, 92.1, 96.7, 97.6, 97.0, 93.9
После 87.1, 84.1, 81.3, 77.0, 86.0, 82.9, 83.0, 85.5, 85.2, 84.6
Проверить гипотезу о, том что средний вес пациентов после диеты статистически
меньше веса до диеты
1) Используйте alternative='greater‘
2) alternative=‘less‘
3) 'two-sided'
Объясните полученные результаты p-value для каждого случая

```python
w_before = [92.8 , 95.6, 92.1, 100.6, 96.2, 92.1, 96.7, 97.6, 97.0, 93.9]
w_after = [87.1, 84.1, 81.3, 77.0, 86.0, 82.9, 83.0, 85.5, 85.2, 84.6]

# необходимо использовать t-тест для зависимых выборок
# H0: M1 = M2
# H1: M1 != M2
bartlett(w_before, w_after) # pvalue=0.877623758816363 - дисперсии равны
st.ttest_rel(w_before, w_after, alternative='greater') # pvalue=1.802e-05 - отклоняем Н0 о равенстве средних (веса)
st.ttest_rel(w_before, w_after, alternative='less') # pvalue=0.99 - принимаем нулевую гипотезу (вес не стал больше)
st.ttest_rel(w_before, w_after) # двусторонний тест твой бро
```

В одной группе из 100 больных наблюдалось улучшение у 75, а в другой из 100 больных
среди 69. Оценить с помощью доверительного интервала разность долей больных, у
которых наблюдались улучшения.

```python
# Задача построения доверительных интервалов для долей
alfa = 1.96
delta = 0.75 - 0.69
m1 = 75
m2 = 69
n1=100
n2 = 100
p_common = (m1 + m2)/(n1+n2)
S_delta = np.sqrt((p_common*(1-p_common))*(1/n1 + 1/n2))
p_up = delta + alfa*S_delta
p_down = delta - alfa*S_delta
f'[{round(p_down, 2)}; {round(p_up, 2)}]' # принимаем Н0 о равенстве средних 
```
